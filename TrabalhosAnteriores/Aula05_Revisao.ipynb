## Q5. Pre-Processamento, POS-tagging, Parser

```python
import nltk
import re
nltk.download('maxent_ne_chunker')
nltk.download('words')
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from pprint import pprint

import joblib
teste_tagger = joblib.load('POS_tagger_brill.pkl')
#phrase = 'O rato roeu a roupa do rei de Roma'
#print(teste_tagger.tag(word_tokenize(phrase)))

print(' ')
print('** QUESTAO 2 **\n')
textoq2 = "Está cada vez mais caro abastecer os veículos nos postos de combustível em Palmas. O preço em R$/litro da gasolina, por exemplo, pode chegar a R$5,89 na capital, segundo pesquisa realizada pelo Procon no dia 1o de março. Antes da pandemia o preço do litro do combustível não passava de R$4,67. As constantes altas, anunciadas pela Petrobras, preocupam consumidores e afetam setores da economia."
print(re.findall(r"[Rr]\$[0-9]+", textoq2)) # usar o \ antes de caracteres especiais

print(' ')
print('** QUESTAO 5 **\n')

textoq5 = "Eu mandei você estudar, e não brincar. Você está de castigo!"

print('5.1. Resultado Pre-processamento: Tokenização \n')
# Tokenizando as sentenças do texto
sentencas = sent_tokenize(textoq5)
print('------- Sentencas: -------')
print(sentencas)
print('------- Tokens p/ Sentença: -------')
sent1 = word_tokenize(sentencas[0])
print(sent1)
sent2 = word_tokenize(sentencas[1])
print(sent2)
print("\n")

print(' ')
print('5.2. Resultado Pre-processamento: remocao de stopwords \n')
print('------------  Lista de StopWords - Portuguese -------\n')
stop_words = set(stopwords.words('portuguese')) 
print(stop_words)

sent1_sstopwrds = [w for w in sent1 if not w in stop_words] 
sent2_sstopwrds = [w for w in sent2 if not w in stop_words] 
print(' ')
print(sent1_sstopwrds)
print(sent2_sstopwrds)



print(' ')
print('5.3. Resultado POS-Tagger -------\n')
print(teste_tagger.tag(word_tokenize(textoq5)))


#!python -m spacy download pt_core_news_sm
import spacy 
from spacy import displacy
from collections import Counter
import pt_core_news_sm
nlp = spacy.load("pt_core_news_sm")


print(' ')
print('5.4. Resultado Parser Sintático (SPaCy) \n')  


displacy.render(nlp(str(sentencas[0])), style='dep', jupyter = True, options = {'distance': 180})
displacy.render(nlp(str(sentencas[1])), style='dep', jupyter = True, options = {'distance': 180})
```

```python
print(' ')
print('** QUESTAO 6 **\n')
print(' ')
print('6.1. Resultado NER -------\n')


doc = nlp(textoq2)

print("\nEntidades e Tipos:")
pprint([(X.text, X.label_) for X in doc.ents])
print("\n")

print(' ')
print('------- Apresentação PRESENTACAO GRAFICA DA NER -------\n')    
displacy.render(nlp(str(textoq2)), jupyter=True, style='ent')

```


## Q7. Chunker


```python
from nltk.chunk import RegexpChunkParser 

#criando um padrão para chunks
pattern = 'NP: {<ART>?<ADJ>*<N>}'
pattern1 = 'NP: {<ART>?<ADJ>*<N.*>*}'  # N, NPROP


print(' ')
print('** QUESTAO 7 **\n')
print(' ')

textoq7 = "As constantes altas no preço da gasolina, anunciadas pela Petrobras, preocupam consumidores e afetam setores da economia."

texto_tagged = teste_tagger.tag(word_tokenize(textoq7))
print(texto_tagged)

print(' ')
print('7.1. CHUNKING (em Flat Structure) - NP: {<ART>?<ADJ>*<N>} \n')
print(pattern)
cp = nltk.RegexpParser(pattern)
cs = cp.parse(texto_tagged)

print(cs)
print("\n")

print('7.2. CHUNKING (em Flat Structure) - NP: {<ART>?<ADJ>*<N.*>*} \n')
print(pattern1)
cp = nltk.RegexpParser(pattern1)
cs = cp.parse(texto_tagged)

print(cs)
print("\n")






```
