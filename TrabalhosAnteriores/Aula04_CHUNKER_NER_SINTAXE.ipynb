## 1. Chunking

```python
import nltk
nltk.download('maxent_ne_chunker')
nltk.download('words')
from nltk.tokenize import word_tokenize
from nltk.tokenize import TweetTokenizer 
from nltk.tag import pos_tag
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint
from nltk.chunk.regexp import ChunkString, ChunkRule, ChinkRule 
from nltk.tree import Tree 
from nltk.chunk import RegexpChunkParser 

texto1 = "@USER Canada doesnâ€™t need another CUCK! We already have enough #LooneyLeft #Liberals f**king up our great country!!! #Qproofs #TrudeauMustGo"
texto2 = "@USER @USER go home youâ€™re drunk @USER #MAGA #Trump2020 ðŸ‘ŠðŸ‡ºðŸ‡¸ðŸ‘Š URL !!!"
texto = "European authorities fined Google a record $5.1 billion on Wednesday for abusing its power in the mobile phone market and ordered the company to alter its practices"

print(' ')
print('------- PRE-PROCESSAMENTO DE SENTENÃ‡AS -------\n')

def preprocess(sent):
    #sent = word_tokenize(sent)
    tweet_tokenizer = TweetTokenizer()
    sent = tweet_tokenizer.tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent

print(' ')
print('------- SENTENÃ‡A PRE-PROCESSADA -------\n')
sent = preprocess(texto)
print(sent)
print("\n")

#criando um padrÃ£o para chunks
pattern = 'NP: {<DT>?<JJ>*<NN>}'
pattern1 = 'NP: {<DT>?<JJ>*<NN.*>*}'  # NN, NNS, NNP, NNPS
#pattern2 = 'NP: {<DT><NN.*><.*>*<NN.*>}'

#('the', 'DT'), ('beautiful', 'JJ'),('mobile', 'JJ'), ('phone', 'NN')  => pattern ok  / pattern1 ok
#('the', 'DT'), ('mobile', 'JJ'), ('Iphone', 'NNP') => pattern NOK / pattern1 ok
#('the', 'DT'), ('mobile', 'JJ') => pattern NOK / pattern1 ok


print(' ')
print('------- CHUNKING (em Flat Structure) PATTERN-------\n')
print(pattern)
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sent)

print(cs)
print("\n")

print('------- CHUNKING (em Flat Structure) PATTERN #1-------\n')
print(pattern1)
cp = nltk.RegexpParser(pattern1)
cs = cp.parse(sent)

print(cs)
print("\n")

print(' ')
print('------- CHUNKING (em IOB format) -------\n')
#IOB tags tem sido usadas para representar a estrutura de chunks
#I-inside O-outside B-beginning (B Ã© usado somente quando existem tags seguidas de mesmo tipo, sem O tokens entre elas.)

iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)


# ChunkString() inicia com uma flat tree 
#tree = Tree('S', [('the', 'DT'), ('book', 'NN'),('has', 'VBZ'), ('many', 'JJ'), ('chapters', 'NNS')]) 
print(' ')
print('------- CHUNKING / CHINKING RULES -------\n')
tree = Tree('S', sent)
print(sent)
print("\n")
print("\n")

# Inicializando ChunkString() 
chunk_string = ChunkString(tree) 
print ("Chunk String : ", chunk_string) 
  
# Inicializando ChunkRule 
chunk_rule = ChunkRule('<DT>?<JJ>*<NN.*><.*>*<NN.*>', 'chunk determinantes, adjetivos e nomes') 
chunk_rule.apply(chunk_string) 
print ("\nApplied ChunkRule : ", chunk_string) 
  
# Criando uma ChinkRule 
chink_rule = ChinkRule('<VB.*>', 'chink verbs') 
chink_rule.apply(chunk_string) 
print ("\nApplied ChinkRule : ", chunk_string, "\n") 



```

## 2.1. NER usando pacote NLTK - PERSON, GPE e ORGANIZATION


```python

print(' ')
print('------- EXEMPLOS DE NER COM NLTK -------\n')

# NER na sentenca "European authorities fined Google a record $5.1 billion on Wednesday for abusing its power in the mobile phone market and ordered the company to alter its practices"
ne_tree = nltk.ne_chunk(sent)
print(ne_tree)
print(type(ne_tree))
print("\n")

# NER na sentenca = "Mark and John S. Marchall are working at Google."
sentence = "Mark and John S. Marchall are working at Google."
ne_tree = nltk.ne_chunk(pos_tag(word_tokenize(sentence)))
print(ne_tree)
print("\n")

iob_tagged = tree2conlltags(ne_tree)
pprint(iob_tagged)
```


## 2.2 NER usando pacote SpaCy


```python

import spacy 
#!python -m spacy download en_core_web_sm
from spacy import displacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()

print(' ')
print('------- EXEMPLOS DE NER COM SpaCy -------\n')
  
sentence = "O ministro da EducaÃ§Ã£o, Milton Ribeiro, decidiu exonerar o presidente da CoordenaÃ§Ã£o de AperfeiÃ§oamento de Pessoal de NÃ­vel Superior (Capes), Benedito GuimarÃ£es Aguiar Neto. Aguiar, engenheiro que estava no cargo desde janeiro de 2020, confirmou ao GLOBO sua saÃ­da e disse que nÃ£o foi informado sobre os motivos da demissÃ£o. Ele afirmou tambÃ©m que nÃ£o sabe quem assumirÃ¡ a presidÃªncia da Capes, ligada ao MinistÃ©rio da EducaÃ§Ã£o (MEC)."
doc = nlp(sentence) 
print(type(doc))

print("SentenÃ§a: "+sentence)
print("\nEntidades e Tipos:")

for ent in doc.ents: 
    print(ent.text, ent.start_char, ent.end_char, ent.label_) 
print("\n")

sentence = "O ministro da EducaÃ§Ã£o, Milton Ribeiro, decidiu exonerar o presidente da CoordenaÃ§Ã£o de AperfeiÃ§oamento de Pessoal de NÃ­vel Superior (Capes), Benedito GuimarÃ£es Aguiar Neto. Aguiar, engenheiro que estava no cargo desde janeiro de 2020, confirmou ao GLOBO sua saÃ­da e disse que nÃ£o foi informado sobre os motivos da demissÃ£o. Ele afirmou tambÃ©m que nÃ£o sabe quem assumirÃ¡ a presidÃªncia da Capes, ligada ao MinistÃ©rio da EducaÃ§Ã£o (MEC)."    
doc = nlp(sentence)
print("SentenÃ§a: "+sentence)
print("\nEntidades e Tipos:")
pprint([(X.text, X.label_) for X in doc.ents])
print("\n")

# ImpressÃ£o de entidades no formato IOB
# "B" means the token begins an entity, 
# "I" means it is inside an entity, 
# "O" means it is outside an entity, and 
# "" means no entity tag is set.
pprint([(X, X.ent_iob_, X.ent_type_) for X in doc])
```

## 2.3 Manipulando Entidades com SpaCy


```python
texto_artigo="""Peter Strzok, the F.B.I. senior counterintelligence agent who disparaged President Trump in inflammatory
text messages and helped oversee the Hillary Clinton email and Russia investigations, has been fired for violating 
bureau policies, Mr. Strzokâ€™s lawyer said Monday. Mr. Trump and his allies seized on the texts, exchanged during 
the 2016 campaign with a former F.B.I. lawyer, Lisa Page, in assailing the Russia investigation as an illegitimate 'witch hunt.'
Mr. Strzok, who rose over 20 years at the F.B.I. to become one of its most experienced counterintelligence agents,
was a key figure in the early months of the inquiry. Along with writing the texts, Mr. Strzok was accused of sending 
a highly sensitive search warrant to his personal email account. The F.B.I. had been under immense political pressure 
by Mr. Trump to dismiss Mr. Strzok, who was removed last summer from the staff of the special counsel, Robert S. Mueller III. 
The president has repeatedly denounced Mr. Strzok in posts on Twitter, and on Monday expressed satisfaction that he had been sacked."""

print(' ')
print('------- TEXTO LONGO -------\n')
print(texto_artigo)
print("\n")

article = nlp(texto_artigo)
print("No. Entidades= "+str(len(article.ents)))
print("\n")

print(' ')
print('------- Contagem de Entidades por LABEL -------\n')
labels = [x.label_ for x in article.ents]
contagem_labels=Counter(labels)
print(contagem_labels)
print("\n")

print(' ')
print('------- TOP 3 Entidades -------\n')
items = [x.text for x in article.ents]
contagem_entidades=Counter(items).most_common(3)
print(contagem_entidades)
print("\n")

#print(type(article.sents))
sentences = [x for x in article.sents]
#print(sentences[2])

print(' ')
print('------- SENTENCAS DO TEXTO LONGO -------\n')
i=1
for x in sentences: 
    print("S"+str(i)+": "+str(x)) 
    i=i+1
    
print(' ')
print('------- APRESENTACAO GRAFICA DA NER da SENTENÃ‡A#3 -------\n')    
displacy.render(nlp(str(sentences[2])), jupyter=True, style='ent')


```
